#**4. Drop null values because the percentage of null values is 0**

```python
df.dropna(inplace=True)
```

# **5. Univariant Analysis**

```python
# Gender
sns.countplot(data=df, x='Sex')
plt.title('Gender Distribution')
```

```python
# Passenger Class
sns.countplot(data=df, x='Pclass')
plt.title('Passenger Class Distribution')
```

```python
# Age
sns.histplot(df['Age'].dropna(), bins=30, kde=True)
plt.title('Age Distribution')
```

```python
# Fare
sns.histplot(df['Fare'], bins=30, kde=True)
plt.title('Fare Distribution')
```

# **6. Bivariant Analysis**

```python
# Survival by Gender
sns.countplot(data=df, x='Survived', hue='Sex')
plt.title('Survival by Gender')
```

```python
# Survival by Pclass
sns.countplot(data=df, x='Survived', hue='Pclass')
plt.title('Survival by Class')
```

# **7. Multivarient Analysis**

```python
# Correlation
numeric_df = df.select_dtypes(include=[np.number])  # Keep only numeric columns
plt.figure(figsize=(10,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
```

```python
sns.pairplot(df[['Survived', 'Pclass', 'Age', 'Fare']].dropna(), hue='Survived')
```
